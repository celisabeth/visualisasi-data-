```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('/content/Most Popular Programming Languages.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
bar_fig = px.bar(df.mean(numeric_only = True).reset_index(),
                 x = 'index', y = 0,
                 title = 'Rata-rata Popularitas Bahasa Pemrograman',
                 labels = {'index': 'Programming Language', 0: 'Average Percentage (%)'},
                 width =1500 ,height = 600)

bar_fig.show()
```

***Tujuan***
Visualisasi data ini untuk menggambarkan informasi terkait bahasa pemprograman apa saja yang paling sering dipakai atau sedang popular. Hasil visualisasi bertujuan untuk memberikan hasil tren di dunia programming, bahasa pemprograman apa saja yang paling banyak digunakan
